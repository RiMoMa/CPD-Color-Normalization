# Imágenes Lizard,segmentación con StarDist

```python
def crop_white_border(image, image_file):
    # If the image is 'glas9.png', return it without cropping
    if image_file == "glas_9.png":
        return image

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    # Threshold the image to create a binary mask (white background will be 255)
    _, binary = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY)

    # Invert the mask to find the actual image area
    binary = cv2.bitwise_not(binary)
    
    # Find contours of the non-white areas
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if contours:
        # Get the bounding box of the largest contour (assumed to be the image content)
        x, y, w, h = cv2.boundingRect(contours[0])

        # Crop the image using the bounding box
        cropped_image = image[y:y+h, x:x+w]
    else:
        # If no contours are found, return the original image
        print("No contours found for this image.")
        cropped_image = image
    
    return cropped_image

# Carpetas de las imágenes
folder1_path = "/kaggle/input/lizard-test/originals_seg"
folder2_path = "/kaggle/input/lizard-test/normalized_seg"

# Lista de imágenes
image_files = os.listdir(folder1_path)

# Figura
fig, axes = plt.subplots(len(image_files), 2, figsize=(15, len(image_files) * 5))

# Impresión de las imágenes lado a lado
for i, image_file in enumerate(image_files):
    # Carga
    image1 = cv2.imread(os.path.join(folder1_path, image_file))
    image2 = cv2.imread(os.path.join(folder2_path, image_file))

    # A RGB
    image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
    image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

    # Crop the white borders from both images, except for 'glass9.png'
    cropped_image1_rgb = crop_white_border(image1_rgb, image_file)
    cropped_image2_rgb = crop_white_border(image2_rgb, image_file)

    # Lado a lado
    axes[i, 0].imshow(cropped_image1_rgb)
    axes[i, 0].axis("off")  # Hide the axis
    axes[i, 0].set_title(f"Original: {image_file}")
    
    axes[i, 1].imshow(cropped_image2_rgb)
    axes[i, 1].axis("off")  # Hide the axis
    axes[i, 1].set_title(f"Normalizada: {image_file}")

# Título
fig.suptitle("Comparación lado a lado", fontsize=16)

# Layout
plt.tight_layout()
plt.subplots_adjust(top=0.95)

plt.show()
```

```python
def normalize_images(input_folder, output_folder, target_image_path):
    """
    This function takes a folder of images, applies Vahadane normalization to each image,
    and saves the normalized images to a specified output folder.
    The target image (reference) is passed as a path parameter.

    Args:
    input_folder (str): The path to the folder containing the images to be normalized.
    output_folder (str): The path to the folder where the normalized images will be saved.
    target_image_path (str): The path to the reference target image for normalization fitting.
    """
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get list of image files in the input folder
    image_files = [f for f in os.listdir(input_folder) if f.endswith(('.jpg', '.png', '.jpeg', '.bmp'))]

    # Initialize the Vahadane Normalizer
    norm = VahadaneNormalizer()

    # Load the target image (reference)
    target_image = cv2.imread(target_image_path)
    target_image_rgb = cv2.cvtColor(target_image, cv2.COLOR_BGR2RGB)

    # Fit the normalizer on the target image
    norm.fit(target_image_rgb)

    # Process each image in the input folder
    for image_file in image_files:
        # Construct the full path to the image
        image_path = os.path.join(input_folder, image_file)
        
        # Load the image
        image = cv2.imread(image_path)
        
        # Convert to RGB
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Apply Vahadane normalization
        normalized_image = norm.transform(image_rgb)
        
        # Convert normalized image back to BGR for saving
        normalized_image_bgr = cv2.cvtColor(normalized_image, cv2.COLOR_RGB2BGR)
        
        # Save the normalized image to the output folder
        output_image_path = os.path.join(output_folder, image_file)
        cv2.imwrite(output_image_path, normalized_image_bgr)

    print("Terminado!")
    
# input_folder = '/kaggle/input/lizard-test/input'
# target_image_path = '/kaggle/input/referencias/testA_35.png'
# mother_path = './Normalized'
# os.makedirs(mother_path, exist_ok=True)
# lizard_path = os.path.join(mother_path, "Lizard")
# os.makedirs(lizard_path, exist_ok=True)
# output_folder = os.path.join(lizard_path, "Vahadane")
# os.makedirs(output_folder, exist_ok=True)
# normalize_images(input_folder, output_folder, target_image_path)


input_folder = '/kaggle/input/glas-test/full-glas'
target_image_path = '/kaggle/input/referencias/dpath_67.png'
glas_path = os.path.join(mother_path, "Glas")
os.makedirs(glas_path, exist_ok=True)
output_folder = os.path.join(glas_path, "Vahadane")
os.makedirs(output_folder, exist_ok=True)
normalize_images(input_folder, output_folder, target_image_path)

```

```python
def display_images_side_by_side(folders, image_names, column_titles, reference_image_path):
    # Set up the figure with a custom gridspec
    num_images = len(image_names)
    num_folders = len(folders)

    # Create a figure
    fig = plt.figure(figsize=(15, 3 * (num_images + 1)))

    # Define a gridspec layout: one row for the reference image, and the rest for the images
    gs = gridspec.GridSpec(num_images + 1, num_folders, height_ratios=[1] + [1] * num_images)

    # Load the reference image
    reference_image = cv2.imread(reference_image_path)
    reference_image_rgb = cv2.cvtColor(reference_image, cv2.COLOR_BGR2RGB)

    # Place the reference image in the top-center of the layout
    ax_ref = fig.add_subplot(gs[0, num_folders // 2])
    ax_ref.imshow(reference_image_rgb)
    ax_ref.axis('off')  # Turn off axis for better display
    ax_ref.set_title('Referencia', fontsize=14)

    # For each image
    for i, image_name in enumerate(image_names):
        # For each folder
        for j, folder in enumerate(folders):
            # Load the image
            image_path = os.path.join(folder, image_name)
            image = cv2.imread(image_path)
            
            # Convert image to RGB (since OpenCV loads images in BGR by default)
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            
            # Display the image in the corresponding subplot
            ax = fig.add_subplot(gs[i + 1, j])
            ax.imshow(image_rgb)
            ax.axis('off')  # Turn off axis for better display
            
            if i == 0:  # Add column titles only to the first row (after the reference)
                ax.set_title(column_titles[j])

            # Add image name below each image
            ax.set_xlabel(image_name, fontsize=10, color='black', horizontalalignment='center', verticalalignment='top')

    plt.tight_layout(h_pad=0.5)  # Adjust horizontal padding for better spacing
    plt.show()

# List of folders containing the images
folders = [
    '/kaggle/input/lizard-test/input',
    '/kaggle/input/lizard-test/output',
    '/kaggle/input/other-normalizations/Lizard/Vahadane',
    '/kaggle/input/other-normalizations/Lizard/Macenko',
    '/kaggle/input/other-normalizations/Lizard/Reinhard'
]

# Get the image names from one of the folders (assuming all have the same images)
image_names = os.listdir(folders[0])

# Filter out non-image files (optional)
image_names = [name for name in image_names if name.endswith(('.jpg', '.png', '.jpeg'))]

# Column titles for each folder
column_titles = ['Original', 'CPD', 'Vahadane', 'Macenko', 'Reinhard']

# Path to the reference image
reference_image_path = '/kaggle/input/referencias/testA_35.png'  # Replace with the actual path

# Display the images
display_images_side_by_side(folders, image_names, column_titles, reference_image_path)
```

```python
import glob
# Definir la carpeta del ground truth y los 7 conjuntos de resultados
ground_truth_folder = "/kaggle/input/glas-masks/masks_ground_truth"
result_folders = [
    "/kaggle/input/glas-masks/1229",
    "/kaggle/input/glas-masks/15005",
    "/kaggle/input/glas-masks/GlaS-testA34Norm",
    "/kaggle/input/glas-masks/GlaSMasks-original",
    "/kaggle/input/glas-masks/macenko",
    "/kaggle/input/glas-masks/reinhard",
    "/kaggle/input/glas-masks/vahadane"
]

# Cargar las máscaras del Ground Truth
ground_truth = {}
for file in glob.glob(os.path.join(ground_truth_folder, "*_anno.bmp")):  # Filtramos por _anno.bmp
    # Obtener el ID de la imagen (sin el sufijo _anno)
    image_id = os.path.basename(file).replace('_anno.bmp', '')
    ground_truth[image_id] = cv2.imread(file, cv2.IMREAD_GRAYSCALE)  # Leer la máscara en escala de grises

# Cargar las máscaras generadas de cada uno de los 7 conjuntos de resultados
generated_masks_per_set = {}  # Diccionario para almacenar las máscaras generadas por cada conjunto

for result_folder in result_folders:
    generated_masks = {}
    for file in glob.glob(os.path.join(result_folder, "*_mask.png")):  # Filtramos por _mask.png
        # Obtener el ID de la imagen (sin el sufijo _mask)
        image_id = os.path.basename(file).replace('_mask.png', '')
        generated_masks[image_id] = cv2.imread(file, cv2.IMREAD_GRAYSCALE)  # Leer la máscara generada en escala de grises
    
    # Almacenamos las máscaras de este conjunto de resultados
    generated_masks_per_set[result_folder] = generated_masks

# Ordenar las claves (IDs) para asegurarse de que las comparaciones sean consistentes
sorted_ids = sorted(ground_truth.keys())  # Usamos las claves del ground truth

# Verificar que las imágenes se han cargado correctamente y que ambos diccionarios contienen las mismas claves
for image_id in sorted_ids:
    if image_id not in ground_truth:
        print(f"Error: No se encontró ground truth para {image_id}")
    for result_folder, generated_masks in generated_masks_per_set.items():
        if image_id not in generated_masks:
            print(f"Error: No se encontró máscara generada para {image_id} en {result_folder}")

# Función para calcular el Dice Score
def dice_score(gt_mask, pred_mask):
    intersection = np.sum(gt_mask & pred_mask)  # Intersección de las máscaras
    union = np.sum(gt_mask) + np.sum(pred_mask)  # Unión de las máscaras
    return 2 * intersection / union if union != 0 else 1.0  # Retornar el Dice Score

# Calcular los Dice Scores para todas las imágenes ordenadas para cada conjunto de resultados
dice_scores_per_set = {}

for result_folder, generated_masks in generated_masks_per_set.items():
    dice_scores = []
    for image_id in sorted_ids:
        gt_mask = ground_truth[image_id]
        pred_mask = generated_masks[image_id]
        score = dice_score(gt_mask, pred_mask)
        dice_scores.append(score)

    # Almacenar los Dice Scores por conjunto
    dice_scores_per_set[result_folder] = dice_scores

# Mostrar los resultados
for result_folder, dice_scores in dice_scores_per_set.items():
    #print(f"\nResultados para el conjunto de resultados: {result_folder}")
    # for image_id, score in zip(sorted_ids, dice_scores):
    #     print(f"Dice Score para {image_id}: {score}")
    
    # Calcular el valor promedio y la desviación estándar de los Dice Scores
    average_score = np.mean(dice_scores)
    std_dev = np.std(dice_scores)

    print(f"Dice Score promedio para {result_folder}: {average_score}")
    print(f"Desviación estándar para {result_folder}: {std_dev}")
    print("")
```

```python
# Calcular los Dice Scores para cada conjunto de resultados (esto ya lo tienes en `dice_scores_per_set`)

# Preparar los datos para el boxplot
dice_scores_list = []
labels = []

for result_folder, dice_scores in dice_scores_per_set.items():
    dice_scores_list.append(dice_scores)
    labels.append(result_folder.split("/")[-1])  # Extraemos solo el nombre de la carpeta

# Crear el boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(dice_scores_list, labels=labels)

# Añadir título y etiquetas
plt.title('Distribución de los Dice Scores por conjunto de resultados')
plt.xlabel('Conjunto de Resultados')
plt.ylabel('Dice Score')

# Mostrar el gráfico
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x para mejor visualización
plt.tight_layout()
plt.show()
```
